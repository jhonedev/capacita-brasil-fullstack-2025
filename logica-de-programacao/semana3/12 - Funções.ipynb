### 1. O que é uma função e por que são úteis?

Uma função é um bloco de código organizado e reutilizável que é usado para realizar uma única ação ou tarefa específica. Funções fornecem uma maneira de modularizar seu código, tornando-o mais organizado, reutilizável e fácil de ler e manter.

#### Utilidade das Funções:

1. `Reutilização de Código`: Em vez de escrever o mesmo código várias vezes, você pode encapsulá-lo em uma função e chamar essa função sempre que a ação associada for necessária.
   
2. `Facilidade de Manutenção`: Alterações feitas dentro de uma função são refletidas em todas as chamadas à função, tornando mais fácil atualizar o código e testar.
   
3. `Decomposição ou modularidade`: Funções permitem dividir programas grandes em componentes menores, facilitando o desenvolvimento e o teste.
   
4. `Abstração`: Você pode esconder detalhes complexos dentro de uma função, permitindo que outros desenvolvedores entendam o código mais facilmente.

### 2. Como Declarar uma Função

Em Python, você usa a palavra-chave `def` para declarar ou definir uma função. A sintaxe geral é:

```python
def nome_da_funcao(parametro1, parametro2, ...):
    # Corpo da função
    # Código para realizar uma tarefa específica
    # Pode incluir declarações, operações e estruturas de controle
    return valor_retornado  # Opcional, apenas se a função deve retornar um valor    
```

A segir uma explicação da sintaxe de uma função em Python:

1. **`def`**: A palavra-chave `def` é usada para começar a definição de uma função. Ela é seguida pelo nome da função.

2. **`nome_da_funcao`**: O nome da função é um identificador que você escolhe para a função. Deve seguir as regras de nomenclatura de variáveis em Python. O nome da função é como você chamará a função posteriormente para executá-la.

3. **Parâmetros (parametro1, parametro2, ...)**: Os parâmetros são variáveis ​​que a função aceita como entrada e são separados por vírgula. Eles estão entre parênteses e podem ser usados ​​dentro do corpo da função. Uma função pode aceitar zero ou mais parâmetros. Os parâmetros são opcionais, ou seja, você pode ter uma função sem parâmetros.

4. **`:` (dois pontos)**: O dois pontos indica o início do bloco de código da função. Todo o código dentro da função deve estar indentado (recuado) para a direita em relação ao `:`.

5. **Corpo da função**: O corpo da função é onde você escreve o código que realiza uma tarefa específica quando a função é chamada. Este é o lugar onde você coloca a lógica da função.

6. **`return` (opcional)**: A instrução `return` é usada para especificar o valor que a função deve retornar quando for chamada. É opcional, ou seja, você pode ter funções que não retornam valores. Se a função tiver uma instrução `return`, ela interromperá a execução da função e retornará o valor especificado. Se não houver uma instrução `return`, a função retornará implicitamente `None`.

### 3. Indentação em funçoes

Em uma função as suas linhas de código devem estar todas indentadas para que o python entenda que são instruções relativas da função. Todas as linhas de código dento da função devem ter o mesmo nível.

Por exemplo:

```python
def calcular_media(notas):
    total = sum(notas)
    media = total / len(notas)
    return media
```


### 4. Passagem de argumentos para uma função

**Diferença entre Parâmetros e Argumentos**

Embora os termos "parâmetros" e "argumentos" sejam frequentemente usados de forma intercambiável, eles têm diferenças sutis:

- `Parâmetros` são as variáveis que são definidas na declaração da função. Eles atuam como espaços reservados para os valores que a função vai receber quando for chamada.

- `Argumentos` são os valores reais que são passados para uma função quando ela é chamada. Esses valores são atribuídos aos parâmetros da função para serem usados no seu bloco de código.



Existem várias maneiras de passar argumentos para uma função:

**Posicionais**: Os argumentos são passados na mesma ordem dos parâmetros definidos.
```python
def calcular_desconto(valor_mensalidade, percentual_desconto):  
    
    desconto = valor_mensalidade * (percentual_desconto / 100)
    valor_com_desconto = valor_mensalidade - desconto    
    
    return valor_com_desconto
    
desconto = calcular_desconto(750, 10)
print (desconto)
```
   
**Nomeados**: Os argumentos são passados explicitamente para cada parâmetro.
```python
def calcular_desconto(valor_mensalidade, percentual_desconto):
    
    desconto = valor_mensalidade * (percentual_desconto / 100)
    valor_com_desconto = valor_mensalidade - desconto   
    
    return valor_com_desconto
    
desconto = calcular_desconto(valor_mensalidade = 750, percentual_desconto = 10)
print (desconto)
```

**Valores Default**: Parâmetros podem ter valores padrão, tornando-os opcionais.
```python
def calcular_desconto(valor_mensalidade, percentual_desconto = 10):
    
    desconto = valor_mensalidade * (percentual_desconto / 100)
    valor_com_desconto = valor_mensalidade - desconto    
    
    return valor_com_desconto
    
desconto = calcular_desconto(valor_mensalidade = 750)
print (desconto)
```


### Exemplo: Função que Aceite Dois Números como Parâmetros e Retorne a Soma deles


```python
def somar_numeros(num1, num2):
    return num1 + num2

# Testando a função
resultado = somar_numeros(5, 10)
print(f"A soma é {resultado}")  # Saída: A soma é 15
```

>Neste exemplo, `num1` e `num2` são parâmetros da função `somar_numeros`. Quando chamamos a função com `somar_numeros(5, 10)`, os valores `5` e `10` são os argumentos que são passados para `num1` e `num2`, respectivamente. A função então retorna a soma desses números.

#### 1. Cálculo da Média de Notas

**Entrada**: Uma lista de notas  
**Processo**: Calcula a média das notas  
**Saída**: Média das notas

```python
def calcular_media(notas):
    soma = sum(notas)
    media = soma / len(notas)
    return media

# Chamada da função
notas_aluno = [85, 90, 78, 92]
media = calcular_media(notas_aluno)
print(f"Média das notas: {media}")
```

#### 2. Verificar Aprovação

**Entrada**: Média das notas  
**Processo**: Verifica se a média é maior ou igual a 60  
**Saída**: Status da aprovação

```python
def verificar_aprovacao(media):
    if media >= 60:
        return "Aprovado"
    else:
        return "Reprovado"

# Chamada da função
w_media = 30
status = verificar_aprovacao(w_media)
print(f"Status: {status}")
```

#### 3. Calcular Percentual de Presença

**Entrada**: Total de aulas e número de aulas assistidas  
**Processo**: Calcula o percentual de presença  
**Saída**: Percentual de presença

```python
def calcular_presenca(total_aulas, aulas_assistidas):
    presenca = (aulas_assistidas / total_aulas) * 100
    return presenca

# Chamada da função
total = 20
assistidas = 18
presenca = calcular_presenca(total, assistidas)
print(f"Percentual de presença: {presenca}%")
```

#### 4. Calcular Nota Final

**Entrada**: Notas das provas, trabalhos e participação  
**Processo**: Calcula a nota final com base nos pesos  
**Saída**: Nota final

```python
def calcular_nota_final(prova, trabalho, participacao):
    nota_final = (prova * 0.5) + (trabalho * 0.3) + (participacao * 0.2)
    return nota_final

# Chamada da função
nota = calcular_nota_final(80, 90, 70)
print(f"Nota Final: {nota}")
```

#### 5. Contar Aprovados e Reprovados

**Entrada**: Uma lista de médias dos alunos  
**Processo**: Conta o número de aprovados e reprovados  
**Saída**: Número de aprovados e reprovados

```python
def contar_status(alunos_media):
    aprovados = 0
    reprovados = 0
    for media in alunos_media:
        if media >= 60:
            aprovados += 1
        else:
            reprovados += 1
    return aprovados, reprovados

# Chamada da função
medias = [85, 40, 77, 90, 55]
aprovados, reprovados = contar_status(medias)
print(f"Aprovados: {aprovados}, Reprovados: {reprovados}")
```
