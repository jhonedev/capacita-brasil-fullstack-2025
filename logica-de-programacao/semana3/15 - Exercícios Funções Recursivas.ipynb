Analise o código Python a seguir e determine o valor de saída do algoritmo observando os escopos local e global das variáveis.

```python
a = 30

def modifica_variaveis():
    global a
    a = 15
    b = 45

modifica_variaveis()
b = a + 10
print (b)
```


<details>
  <summary>Clique para verificar a resposta</summary>

<br>

Após a execução do programa, o valor de `a` será `15` e o valor de `b` será `25`. O comando `global` dentro da função `modifica_variaveis` indica que a variável `a` mencionada na função refere-se à variável `a` definida no escopo global. Assim, quando alteramos o valor de `a` para `15` dentro da função, estamos modificando a variável global `a`. A variável `b` definida dentro da função é local e seu escopo termina com a função; portanto, a atribuição de `b = a + 10` fora da função cria ou modifica a variável global `b`, utilizando o valor atualizado de `a`.


### Exercício 5
Rastrear a execução de função `potencia_recursiva(2, 3)` do exercício anterior. Esta tabela explicitará as `chamadas da função` e seus `retornos`, detalhando o valor de `base`, `expoente`, e a `ação realizada`.

Obs: o expoente não pode ser negativo

```python
def potencia(base, expoente):
    # Caso base: qualquer número elevado a 0 é 1
    if expoente == 0:
        return 1
    # Passo recursivo: base^expoente = base * base^(expoente-1)
    else:
        return base * potencia(base, expoente - 1)
```
