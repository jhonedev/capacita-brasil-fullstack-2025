### Definição de Estrutura de Dados

Uma estrutura de dados é uma maneira específica de organizar e armazenar dados em um computador para que possam ser acessados e modificados de maneira eficiente. Essencialmente, as estruturas de dados fornecem os meios para agrupar e organizar informações de acordo com as relações entre elas.

1. **Estruturas de Dados Básicas em Python**:
   - **Inteiro (`int`)**: Representa números inteiros.
   - **String (`str`)**: Representa sequências de caracteres.
   - **Float (`float`)**: Representa números de ponto flutuante (decimais).
   - **Booleano (`bool`)**: Representa valores de verdadeiro ou falso.
   - **Lista (`list`)**: Representa uma coleção ordenada de itens, que podem ser de qualquer tipo.
   - **Dicionário (`dict`)**: Representa uma coleção de pares chave-valor.
   - **Conjunto (`set`)**: Repesenta uma coleção não ordenada de elementos únicos



**Definição de Objetos:**
As estruturas de dados são objetos. Um objeto é uma coleção de dados e funções que operam nesses dados. De uma forma mais simples, um objeto é como uma "caixa" que contém dados e meios para os manipular(operações, funções ou métodos).

- Tipos de objetos
    - Estruturas de Dados: inteiro, string, float, booleano, lista, dicionário, set, etc

    - Outros tipos:
         - conexões com arquivos (objeto file)
         - conexões com banco de dados (objeto connect)
         - conexões com internet (objeto http)
         - etc

## Estrutura de dados Strings


**Definição:**
Objetos do tipo `string` são sequências de caracteres e são usadas para representar texto em Python.

A representação (literal) de `string` na linguagem Python é entre aspas simples (' ') ou aspas duplas (" ").

**Literais que representam tipo de dados `string`**
```python
"Olá, mundo!"
```


```python
'Python'
```

```python
''' Brasília é
a capital do Brasil
'''
```

Esses literais podem ser usados diretamente em expressões ou para inicializar `variáveis` do tipo `string`.

```python
# Inicializando uma variável com um literal string
pais = 'Brasil'

pais, estado, cidade = 'Brasil', 'Ceará', 'Fortaleza'
```

**Descrição dos Operadores e funções com `string`:**


1. **Concatenação (`+`)**:

     Combina duas strings em uma única string.
     
     `Sintaxe`: string1 + string2
    
  ```python
  saudacao = "Olá"
  nome = "Maria"
  completo = saudacao + ", " + nome
  print(completo)
  ```
      
    Resultado:
  ```python        
  Olá, Maria  
  ```
      
  > Observação: Ambas as partes devem ser strings, ou você obterá um erro de tipo.

2. **Repetição (`*`)**:

     Repete a string um número especificado de vezes.
    
     `Sintaxe`: string * n
    
  ```python
  texto = "A"
  repetido = texto * 3
  print(repetido)
  ```
      
    Resultado:
  ```python
  AAA
  ```
      
  > Observação: `n` deve ser um número inteiro.
      

3. **Indexação (`[index]`)**:

     Obtém o caractere em um índice específico.
     
     `Sintaxe`: string[index]
     
  ```python
  palavra = "exemplo"
  letra = palavra[3]
  print (letra)
  ```
    Resultado:
  ```python
  m
  ```

4. **Fatiamento (`[start:end]`)**:

    Obtém uma substring usando índices de início e fim.
    
    >   O fatiamento é feito usando a notação `string [inicio:fim] `, onde início é o índice inicial e fim é o índice onde o fatiamento deve parar, mas é **exclusivo** (`o elemento no índice fim não é incluído na sublista`).
    
  ```python
  palavra = "exemplo"
  fatia = palavra[1:4]
  print(fatia)
  ```
      
    Resultado:
  ```python        
  xem
  ```
    
5. **Comprimento (`len()`)**:

    Retorna o comprimento da string.
    
    `Sintaxe`: `len(string)`
    
  ```python
  palavra = "exemplo"
  tamanho = len(palavra)
  print(tamanho)
  ```
      
    Resultado:
  ```python        
  7
  ```
    
6. **Encontrar (`find`)**:

    Localiza a primeira ocorrência de uma substring e retorna o índice.
    
    `Sintaxe`: `string.find(substring)`
    
  ```python
  palavra = "exemplo"
  posicao = palavra.find("e")
  print(posicao)
  ```
      
    Resultado:
  ```python        
  0
  ```
      
  > Observação: Retorna `-1` se a substring não for encontrada.
    
7. **Substituir (`replace`)**:

    Substitui todas as ocorrências de uma substring por outra.
    
    `Sintaxe`: `string.replace(old, new)`
    
  ```python
  palavra = "exemplo"
  nova_palavra = palavra.replace("e", "E")
  print(nova_palavra)
  ```
      
    Resultado:
  ```python        
  ExEmplo
  ```
    
8. **Dividir (`split`)**:

    Divide a string em uma lista de substrings com base em um delimitador.
    
    `Sintaxe`: `string.split(delimiter)`
    
  ```Python
  frase = "um exemplo simples"
  palavras = frase.split(" ")
  print(palavras)
  ```
      
    Resultado:
  ```python        
  ["um", "exemplo", "simples"]
  ```
    
9. **Início com (`startswith`)**:

    Verifica se a string começa com uma determinada substring.
    
    `Sintaxe`: `string.startswith(substring)`
    
  ```python
  palavra = "exemplo"
  comeca_com = palavra.startswith("e")
  print(comeca_com)
  ```
      
    Resultado:
  ```python        
  True
  ```
    
10. **Termina com (`endswith`)**:

    Verifica se a string termina com uma determinada substring.
    
    `Sintaxe`: `string.endswith(substring)`
    
  ```python
  palavra = "exemplo"
  termina_com = palavra.endswith("o")
  print(termina_com)
  ```
      
    Resultado:
  ```python        
  True
  ```

11. **Pertença (in)**:
   Verifica se uma substring está presente na string. Se estiver, a expressão retorna `True`, caso contrário, retorna `False`.

```python
# Entrada
nome = 'ventilador'

# Processo
esta_na_string = "r" in nome

# Saída
print(f"Localizada a substring?: {esta_na_string}")

```
	Resultado:
```python
Localizada a substring? : True
```

**Exemplo 1: Concatenar Nome e Sobrenome de um Funcionário**

`Entrada`
- Solicitar o nome e o sobrenome do funcionário.

`Processo`
- Concatenar o nome e o sobrenome para formar o nome completo.

`Saída`
- Exibir o nome completo do funcionário.


````python
# Entrada
nome = input("Digite o nome do funcionário: ")
sobrenome = input("Digite o sobrenome do funcionário: ")

# Processo
nome_completo = nome + " " + sobrenome   # aqui ocorre a concatenação de strings

# Saída
print(f"Nome completo do funcionário: {nome_completo}")
````


> **Instrução: `input`**
>
> A função `input()` em Python é usada para capturar a entrada do usuário. A função pode exibir uma mensagem que atua como um `prompt` para o usuário fornecer a entrada. A entrada do usuário é então lida como uma string e pode ser armazenada em uma variável para uso posterior. Lembre-se de usar sempre os parêntesis abrindo **`(`** e fechando **`)`** quando usar funções.


> **Cláusula: `f-string`**
>
>O prefixo **`"f"`** usado antes de uma string indica que estamos diante de um `tipo especial de string` conhecida como **"f-string"**. As f-strings são uma forma de `interpolação` de strings, o que significa que `variáveis` ou expressões possam ser inseridas diretamente dentro de strings.
>
> Para usar `f-strings`, você começa a string com a letra `f` ou `F` logo antes das aspas de abertura. Dentro da string, qualquer coisa colocada entre chaves **`{ }`** será avaliada como uma expressão Python.
>
>No exemplo acima, `{nome_completo}` será substituído pelo valor da variável `nome_completo`.

**Exemplo 2: Formatar E-mail Corporativo**

`Entrada`
- Solicitar o primeiro nome e o último nome do funcionário.

`Processo`
- Converter o primeiro nome e o último nome para minúsculas.
- Concatenar o primeiro nome e o último nome com um ponto entre eles.
- Adicionar o sufixo "@empresa.com" para formar o e-mail corporativo.

`Saída`
- Exibir o e-mail corporativo formatado.



````python
# Entrada
primeiro_nome = input("Digite o primeiro nome do funcionário: ")
ultimo_nome = input("Digite o último nome do funcionário: ")

# Processo
primeiro_nome = primeiro_nome.lower()
ultimo_nome = ultimo_nome.lower()
email_corporativo = primeiro_nome + "." + ultimo_nome + "@empresa.com"

# Saída
print(f"E-mail corporativo: {email_corporativo}")
````
